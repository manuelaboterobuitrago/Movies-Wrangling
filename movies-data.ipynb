*Verificar cu√°ntos valores nulos hay en cada columna*

```python
movies.isnull().sum()
```

*Rellenar valores nulos en la columna 'GENRE' con 'Desconocido'*

```python

movies.fillna({'GENRE':'Desconocido' }, inplace=True)
```

*Mostrando el relleno de valores nulos*

```python
movies[movies['GENRE'] == 'Desconocido']
```

*Eliminar espacios extra en los nombres de las columnas*

```python
movies.columns = movies.columns.str.strip()
```

```python
movies.head()
```

*Reemplazar m√∫ltiples caracteres no deseados como \n, \t, y otros*

```python
movies = movies.apply(lambda x: x.str.replace(r'[\n\t\r]', ' ', regex=True) if x.dtype == 'object' else x)
```

*Separando el director de los actores en la columna stars, se crea una columna respectivamente*

```python
split_stars = movies['STARS'].str.split('|', expand=True)

movies['DIRECTOR'] = split_stars.iloc[:, :-1].apply(lambda x: ', '.join(x.dropna()), axis=1)
movies['ACTORS'] = split_stars.iloc[:, -1]
```

```python
movies.head()
```

*Eliminando columna 'STARS'*

```python
movies.drop(['STARS'],axis=1, inplace=True)
```

```python
movies.head()
```

*Llenando espacios en blanco de 'ACTORS' y 'DIRECTORS' con Desconocido*

```python
movies['ACTORS'] = movies['ACTORS'].fillna('Desconocido')
movies['DIRECTOR'] = movies['DIRECTOR'].fillna('Desconocido')
```

*Eliminando guiones y par√©ntesis*

```python
movies["YEAR"] = movies["YEAR"].str.replace('(', '')
movies["YEAR"] = movies["YEAR"].str.replace(')', '')
movies["YEAR"] = movies["YEAR"].str.replace('-', '')
```

*Llenando los a√±os que faltan con 0*

```python
movies['YEAR'] = movies['YEAR'].fillna(0)
```

*Llenado 'VOTES' con 0 y 'RATING' tambi√©n con 0*

```python
movies['VOTES'] = movies['VOTES'].fillna('0')
movies['VOTES'] = movies['VOTES'].str.replace(',', '')
movies['VOTES'] = movies['VOTES'].astype(int) 
```

```python
movies['RATING'] = movies['RATING'].fillna(0.0)
```

```python
movies['RunTime'] = movies['RunTime'].fillna(0.0)
```

```python
movies.isnull().sum()
```

*Guardando el dataset limpio*

```python
movies.to_csv('movies_cleaned.csv', index=False)
```
